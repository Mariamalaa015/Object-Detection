#Importing Libaries

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

#Using Configuration files and frozen model with datasets

```python
config_file='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
```

```python
frozen_model ='frozen_inference_graph.pb'
```

#Creating a detection model 

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

#Appending Class labels from dataset to a list

```python
classlabels=[]
file_name='labels.txt'
with open(file_name,'rt') as fpt:
    classlabels = fpt.read().rstrip('\n').split('\n')
```

#Customizing Model

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

# read an image

```python
image = cv2.imread('test2.jpg')
```

```python
plt.imshow(image)
```

#Changing BGR TO RGB

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex,confidence,bbox = model.detect(image,confThreshold=0.5)
```

#Printing the output

```python
print(ClassIndex)
```

```python
print(classlabels[2],classlabels[0])
```

#Putting Text on the picture

```python
font_scale=1.5
font = cv2.FONT_HERSHEY_PLAIN
for classInd , conf , boxes in zip(ClassIndex.flatten(),confidence.flatten(),bbox):
    cv2.rectangle(image,boxes,(255,0,0),2)
    cv2.putText(image,classlabels[classInd - 1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=3)
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

#Video Demo

```python
cap = cv2.VideoCapture("test.mp4")
```

```python
if not cap.isOpened():
    cap=cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("can't open file")
font_scale=3
font = cv2.FONT_HERSHEY_PLAIN
while True:
    ret,frame=cap.read()
    ClassIndex,confidence,bbox = model.detect(frame,confThreshold=0.55)
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for classInd , conf , boxes in zip(ClassIndex.flatten(),confidence.flatten(),bbox):
            if (classInd < 80):
                cv2.rectangle(image,boxes,(255,0,0),2)
                cv2.putText(frame,classlabels[classInd - 1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=3)
    cv2.imshow("Object Detection Tutorial",frame)
    if cv2.waitKey(2) & 0xFF == ord ('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
